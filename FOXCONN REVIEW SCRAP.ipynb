<h1> Scraping details about FOXCONN COMPANY REVIEWS from INDEED.com using BeautifulSoup <h1>

<h3> Importing all the required libraries <h3

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from requests import get
from requests.exceptions import RequestException
import csv
from datetime import datetime
from urllib.request import urlopen as uReq
import re
```

```python
page_ur = "https://www.indeed.com/cmp/Foxconn/reviews?fcountry=ALL"
uClient1 = uReq(page_ur)
page_soup1 = soup(uClient1.read(), "html.parser")
```

<h3> Preparing all the links to be scraped and combining them into one lists<h3>

```python
links = []

links.append(page_ur)
for i in range(20,1630,20):
    links.append("https://www.indeed.com/cmp/Foxconn/reviews?fcountry=ALL&start=%d" %i)
 

    
```

```python
links
len(links)
```

<h2> This is the main script thats going to scrap all the data required from the 82 web pages<h2>

<h4> In this script fRealValue1 list contains the reviews content and fRealValue contains all other information<h4> 

```python
fRealValue1 = []
fRealValue = [] 
for i in range(0,82):
    page_url = links[i]
    uClient = uReq(page_url)
    page_soup = soup(uClient.read(), "html.parser")
    uClient.close()
    k=0
    j=0
    for k in page_soup.find_all("div", {"class":"cmp-review-description"}):
        fRealValue1.append(k.text.strip())
    for j in page_soup.find_all("div", {"class":"cmp-review-subtitle"}):
        fRealValue.append(j.text.strip().split("\xa0"))
```

```python
fRealValue
my_df = pd.DataFrame(fRealValue, columns=['Designation','sd','Data'])
my_df
```
